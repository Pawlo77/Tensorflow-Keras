```python
import matplotlib.pyplot as plt
```

```python
def plot(distrib):
    plt.hist(distrib, bins=100)
    plt.show()
```

```python
from scipy.stats import randint

rand_distrib = randint(1, 100).rvs(1000, random_state=42)
plot(rand_distrib)
```

```python
from scipy.stats import geom

gem_distrib = geom(0.5).rvs(1000, random_state=42)
plot(gem_distrib)
```

```python
from scipy.stats import expon

exp_distrib = expon(scale=1).rvs(1000, random_state=42)
plot(exp_distrib)

# best when we know (more or less) what scale should be
```

```python
from scipy.stats import reciprocal

reci_distrib = reciprocal(1, 1000).rvs(1000, random_state=42)
plot(reci_distrib)

# useful when we don't have a clue of what the target scale is
```
